```python
from sklearn.preprocessing import OrdinalEncoder

print(train_df["education_level"].unique())
education_levels = [
  "High School",
  "Bachelor's",
  "Master's",
  "PhD",
  "Other"
]
print("Ordered Levels: ", education_levels)

ord_encoder = OrdinalEncoder(categories=[education_levels])
train_df["education_level_ordered"] = ord_encoder.fit_transform(train_df["education_level"].values.reshape(-1, 1))
print(train_df["education_level_ordered"].value_counts())
```
